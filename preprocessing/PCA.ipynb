#Principal Component Analysis

The following code below allows combines the standard import routine, along with a Principal Component Analysis of the data.  

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline

# Import the datasets via the read_json method from pandas
pizza_data_train = pd.read_json('../data/train.json')
pizza_data_test = pd.read_json('../data/test.json')

# Analyze the size of the datasets
print "Shape of training data:" + str(pizza_data_train.shape)
print "Shape of test data:" + str(pizza_data_test.shape)
```

```python
from matplotlib import pyplot as plt
import sklearn as sk
from sklearn.decomposition import TruncatedSVD
import pandas as pd
import numpy as np
%matplotlib inline
```

```python
reduced_training_set = pizza_data_train.drop(pizza_data_train.columns[[0,1,2,3,4,6,7,8,22,23,28,29]], 1)
target = pizza_data_train[pizza_data_train.columns[[22]]]

reduced_training_set = (reduced_training_set - reduced_training_set.mean())/(reduced_training_set.max()- reduced_training_set.min())
```

```python
pcaobject = TruncatedSVD(n_components=np.int(3))
Xt = pcaobject.fit_transform(reduced_training_set.values)
```

```python
Xt_Data = pd.DataFrame(Xt)
Xt_Data = Xt_Data.join(target)

pc1 = Xt_Data[Xt_Data.requester_received_pizza == False]
pc1 = pc1.drop("requester_received_pizza", 1)
pc2 = Xt_Data[Xt_Data.requester_received_pizza == True]
pc2 = pc2.drop("requester_received_pizza", 1)

plt.subplots(figsize=(12,6))
plt.title("PCA Analysis on Pizza Data", fontsize=16)
plt.ylabel("Principal component 2", fontsize=12)
plt.xlabel("Principal component 1", fontsize=12)
s1 = plt.scatter(pc1.values[:,0], pc1.values[:,1], c="red")
s2 = plt.scatter(pc2.values[:,0], pc2.values[:,1], c="blue")
plt.legend((s1, s2), ("PC1", "PC2"),loc='best', 
           borderpad=0.3, shadow=False, markerscale=1.4, scatterpoints=1, fontsize=16)
plt.show()
```
